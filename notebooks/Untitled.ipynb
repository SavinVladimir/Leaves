```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
plt.figsize=(12,8)
```

```python
data = pd.read_csv('../data/out.csv')
```

```python
data.head()
```

```python
sns.displot(x='length',kde=True, data=data)
```

```python
sns.displot(x='width',kde=True, data=data)
```

```python
sns.displot(x='length', hue='width', kde=True, data=data);
```

```python
sns.displot(x='length', hue='width',kind="kde",fill=True,data=data);
```

```python
sns.kdeplot(x='length', cut=0, data=data)
```

```python
sns.kdeplot(x='width', cut=0, data=data)
```

```python
sns.displot(x='length',
            y='width',
            bins=10,
            data=data);
```

```python
sns.jointplot(x='length',
              y='width',
              data=data);
```

```python
plt.scatter(y=data['id'],x=data['width'], marker = '+', color='orange')
```
